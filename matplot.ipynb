# Subplots

```python
import numpy as np
x=np.arange(30,40)
y=np.arange(40,50)
```

```python
plt.subplot(3,3,1)  #(rows,columns,index) for the whole plotting area
plt.plot(x,y,'g')
plt.subplot(3,3,2)
plt.plot(x,y,'r')
plt.subplot(3,3,3)
plt.plot(x,y,'y')
plt.subplot(3,3,4)
plt.plot(x,y,'c')
```

# Sin and Cosin graphs

```python
x=np.arange(0,4*np.pi,0.1) # 3rd value is the step size
y=np.sin(x)
plt.plot(x,y)
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
```

# Pie Chart

```python
labels='Python', 'Java', 'C', "C++"
sizes=[200,100,150,120]
colors='green','blue','gold','lightskyblue'
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct="%1.1f%%", shadow= True)

# plt.axis('equal')
```

```python
slices=[7,2,2,13]
act=['sleep','play','work','eat']
cols=['c','m','r','b']
plt.pie(slices,labels=act,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),
      autopct="%1.1f%%" )
plt.show()
```
